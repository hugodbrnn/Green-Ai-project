```python
import pandas as pd

URL = "https://odre.opendatasoft.com/api/explore/v2.1/catalog/datasets/eco2mix-regional-tr/exports/csv"


df = pd.read_csv(URL, sep=";")

# Filtre Occitanie
df_occ = df[df["libelle_region"] == "Occitanie"]

# ‚ö†Ô∏è Enlever les lignes o√π la consommation est NaN
df_occ = df_occ.dropna(subset=["consommation"])

# Prendre la derni√®re valeur r√©elle
latest = df_occ.sort_values("date_heure", ascending=False).iloc[0]

heure = latest["date_heure"]
consommation = latest["consommation"]

print("üîå  DEMANDE D'√âNERGIE EN OCCITANIE")
print("--------------------------------------")
print(f"üïí Heure : {heure}")
print(f"‚ö° Consommation : {consommation} MW")
```
